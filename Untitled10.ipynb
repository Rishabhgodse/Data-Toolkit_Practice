Difference between Deep Copy and Shallow Copy
a. Shallow Copy
Definition: In a shallow copy, a new object is created, but references to the original object's elements are copied, not the actual data. This means that changes made to nested or mutable elements in the copied object will affect the original object because both objects share the same references.

Behavior: Copies the structure, not the actual elements inside mutable objects.


import numpy as np

arr = np.array([[1, 2], [3, 4]])
shallow_copy = arr.view()  # Shallow copy using view()

shallow_copy[0, 0] = 99
print("Original Array:", arr)
Key Point: In the case of shallow copies, changes to the shallow copy's elements affect the original array.
b. Deep Copy
Definition: In a deep copy, a new object is created, and all elements are copied recursively, including nested and mutable elements. Changes made to the deep copy do not affect the original object because both objects are completely independent of each other.

Behavior: Copies the actual data, not the references.


import numpy as np

arr = np.array([[1, 2], [3, 4]])
deep_copy = arr.copy()  # Deep copy using copy()

deep_copy[0, 0] = 99
print("Original Array:", arr)
Key Point: In a deep copy, changes to the deep copy do not affect the original array.

Summary:

Shallow Copy:

Copies references to the original elements.
Changes to the copied object can affect the original if the elements are mutable.
Example: arr.view() in NumPy.
Deep Copy:

Creates a completely new object with independent data.
Changes to the copied object do not affect the original.
Example: arr.copy() in NumPy.

Determine Quadrants: Based on the sign of x and y, classify the points into four quadrants:
Quadrant I: x > 0 and y > 0
Quadrant II: x < 0 and y > 0
Quadrant III: x < 0 and y < 0
Quadrant IV: x > 0 and y < 0

Code Implementation:

import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

# Step 1: Generate random data for x and y coordinates
x = np.random.randn(100)  # 100 random numbers from a normal distribution
y = np.random.randn(100)

# Step 2: Determine the quadrant for each point based on x and y values
quadrants = []
for i in range(len(x)):
    if x[i] > 0 and y[i] > 0:
        quadrants.append('Quadrant I')
    elif x[i] < 0 and y[i] > 0:
        quadrants.append('Quadrant II')
    elif x[i] < 0 and y[i] < 0:
        quadrants.append('Quadrant III')
    else:
        quadrants.append('Quadrant IV')

# Step 3: Create a DataFrame with x, y, and quadrant information
df = pd.DataFrame({
    'X': x,
    'Y': y,
    'Quadrant': quadrants
})

# Step 4: Create the scatter plot using Seaborn
plt.figure(figsize=(8, 6))
sns.scatterplot(data=df, x='X', y='Y', hue='Quadrant', palette='Set1', s=100, marker='o')

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Quadrant-wise Scatter Plot')

# Display the legend
plt.legend(title='Quadrants')

# Show the plot
plt.show()

x = np.random.randn(100) and y = np.random.randn(100) generate 100 random points for x and y coordinates from a normal distribution.
Determining Quadrants:

Creating the DataFrame:

We combine x, y, and the corresponding quadrant labels into a pandas DataFrame.
